```python
import os
import shutil

# Paths
source_root = "Palms_pano_plus"
target_root = "Pano_Json_BE_E2"

# Walk through each building
for building in os.listdir(source_root):
    building_path = os.path.join(source_root, building)
    
    if not os.path.isdir(building_path):
        continue  # Skip non-directory files
    
    # Walk through each session within building
    for session in os.listdir(building_path):
        session_path = os.path.join(building_path, session)
        
        images_path = os.path.join(session_path, "images")
        if not os.path.isdir(images_path):
            continue  # Skip if no images folder
        
        # Target directory for the json files
        target_images_path = os.path.join(target_root, building, session, "images")
        os.makedirs(target_images_path, exist_ok=True)
        
        # Copy all image_*.json files
        for file in os.listdir(images_path):
            if file.endswith(".json") and file.startswith("image_"):
                src_file = os.path.join(images_path, file)
                dst_file = os.path.join(target_images_path, file)
                
                shutil.copy2(src_file, dst_file)
                print(f"Copied {src_file} → {dst_file}")

print("\n✅ JSON extraction complete.")
```
